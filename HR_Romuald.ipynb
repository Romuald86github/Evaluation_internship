```python
def mann_whitney_test(df, columns_to_test, target_column='Attrition'):
    results = []

    for column in columns_to_test:
        # Extract data for 'Yes' and 'No' Attrition
        attrition_yes = df[df[target_column] == 'Yes'][column]
        attrition_no = df[df[target_column] == 'No'][column]

        # Perform Mann-Whitney U test
        statistic, p_value = mannwhitneyu(attrition_yes, attrition_no)

        # Append results to the list
        results.append({
            'Column': column,
            'Mann-Whitney U statistic': statistic,
            'P-value': p_value
        })

    return pd.DataFrame(results)
```
